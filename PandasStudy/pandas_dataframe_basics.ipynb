# Pandas Data Frame

## It's a group of series object that share the same index.

```python
## import numpy and pandas 
import numpy as np
import pandas as pd
```

```python
# create a 3x3 array with random integers between 1 and 10 (numpy array to pandas dataframe)
np.random.seed(101)
array=np.random.randint(low=1,high=10,size=(3,3))
index=['row1','row2','row3']
columns=['col1','col2','col3']
df=pd.DataFrame(data=array,index=index,columns=columns)
df
```

```python
# information about the data frame
df.info()
```

```python
## read csv file using pandas data frame 
df1=pd.read_csv(r'.\DataStore\tips.csv')
df1
```

```python
# to see the column name 
df1.columns #this is an attribute not a method so no () needed
```

```python
# to see the index of the dataframe
df1.index
```

```python
# to see first couple of rows 
df1.head()# default is 5 rows but you can specify the number of rows you want to see by passing an integer value inside the head method
```

```python
# to see last couple of rows
df1.tail()# default is 5 rows but you can specify the number of rows you want to see by passing an integer value inside the tail method
```

```python
# to see the statistical summary of the dataframe
df1.describe()
```

```python
# to transpose the metrix 
df1.describe().transpose()
```

## Retrive Information from Data Frame 

#### Retrive information based on the column 

```python
df1['total_bill'] # to see a particular column pass the column name inside the square brackets
```

```python
## to retrieve multiple columns
column_list=['total_bill','tip']
df1[column_list] # pass the list of column names inside the square brackets
```

```python
# operating on the columns 
df1['total_bill']*2
```

```python
## we can do similar operations like addition,subtraction,division etc similar like numpy arrays 
df1['percentage_of_tips']=np.round((df1['tip']/df1['total_bill'])*100,2)
df1.head()
```

```python
## remove column 
df1.drop(labels='percentage_of_tips',axis=1)
```
